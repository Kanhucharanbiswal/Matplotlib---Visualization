```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [25, 45, 85, 55, 30, 65, 75, 45, 90]
```

```python
colors = ['yellow' if score<50 else 'green' for score in exam_scores]
sizes = [score *2 for score in exam_scores]
plt.scatter(study_hours, exam_scores, c=colors, s=sizes)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
```

```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [25, 45, 85, 55, 30, 65, 75, 45, 90]
```

```python
colors = ['yellow' if score<50 else 'green' for score in exam_scores]
sizes = [score *2 for score in exam_scores]
plt.scatter(study_hours, exam_scores, c=exam_scores, cmap='viridis')
plt.colorbar(label="Score")
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
```

```python
colors = ['yellow' if score<50 else 'green' for score in exam_scores]
sizes = [score *2 for score in exam_scores]
plt.scatter(study_hours, exam_scores, c=exam_scores, cmap='binary')
plt.colorbar(label="Score")
plt.grid(True)
for i in range(len(study_hours)):
    plt.annotate(f'S {i+1}', (study_hours[i], exam_scores[i]))
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study vs Marks")
```
